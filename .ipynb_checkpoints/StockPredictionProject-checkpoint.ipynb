Voici à quoi ressemble une intervalle de l'une de mes datasets, qui sont stockées en Json.

```python
import sys
import json

with open('dataset/dataset_1.txt', 'r') as myfile:
    data = myfile.read()
    dataset1 = json.loads(data)
with open('dataset/dataset_2.txt', 'r') as myfile:
    data = myfile.read()
    dataset2 = json.loads(data)
with open('dataset/dataset_3.txt', 'r') as myfile:
    data = myfile.read()
    dataset3 = json.loads(data)
with open('dataset/dataset_4.txt', 'r') as myfile:
    data = myfile.read()
    dataset4 = json.loads(data)
with open('dataset/dataset_5.txt', 'r') as myfile:
    data = myfile.read()
    dataset5 = json.loads(data)

print(json.dumps(dataset1[0], indent=2))
```

Chaque intervalle contient donc son timestamp, le prix du bitcoin à ce moment et l'orderbook.

Maintenant regardons plus en détails les datasets.

# Bandes de Bollinger

```python
import sys
sys.path.append('./src')
from bollinger import bollinger
import numpy as np


valuestest = prices1[0::100]

def genBands(values):
    bg = bollinger(2, 20)
    bands = []
    for i in range(100):
        if i < 20:
            bg.feed(values[i])
        else:
            bg.feed(values[i])
            bands.append(bg.process())

    bmins = list(map(lambda x: float(x['bmin']), bands))
    bmaxs = list(map(lambda x: float(x['bmax']), bands))
    return { 'bmins': [float('Nan') for x in range(20)] + bmins, 'bmaxs': [float('Nan') for x in range(20)] + bmaxs }

bands = genBands(valuestest)

plt.plot(bands['bmins'], label="bmins", linewidth=0.5)
plt.plot(bands['bmaxs'], label="bmaxs", linewidth=0.5)
plt.plot(valuestest, label="value", linewidth=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

On va maintenant se décider dans quelles intervalles de temps on se place pour trader.

Est-ce que on se place dans des intervalles de l'ordre de grandeur des secondes, des minutes ?
On va comparer des courbes prises sur plusieurs intervalles de temps et les observer sur la même durée.
On va appliquer des bandes de bollinger sur ces courbes pour voir leur volatilité.

```python
#intervalles de 10 secondes
values1 = prices1[0::2][:100]
bands1 = genBands(values1)
#intervalles de 1 minute
values2 = prices1[0::12][:100]
bands2 = genBands(values2)

#intervalles de 5minutes
values3 = prices1[0::60][:100]
bands3 = genBands(values3)

plt.plot(bands1['bmins'], label="bmins", linewidth=0.5)
plt.plot(bands1['bmaxs'], label="bmaxs", linewidth=0.5)
plt.plot(values1, label="value", linewidth=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

plt.plot(bands2['bmins'], label="bmins", linewidth=0.5)
plt.plot(bands2['bmaxs'], label="bmaxs", linewidth=0.5)
plt.plot(values2, label="value", linewidth=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

plt.plot(bands3['bmins'], label="bmins", linewidth=0.5)
plt.plot(bands3['bmaxs'], label="bmaxs", linewidth=0.5)
plt.plot(values3, label="value", linewidth=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

On remarque que les intervalles de 10 secondes sont beaucoup trop courtes pour trade, il se passe n'importe quoi.

Les intervalles de 1 minutes sont intéressantes mais parfois le cours stagne pendant trop longtemps ce qui dérègle les courbes de bollinger.

Les intervalles de 5 minutes semblent être les plus propices aux bandes de bollinger.
